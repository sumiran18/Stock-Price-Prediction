```python
Adj_Close_Price = google_data[['Adj Close']]
```

```python
max(Adj_Close_Price.values), min(Adj_Close_Price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(Adj_Close_Price)
```

```python
scaled_data
```

```python
len(scaled_data)
```

```python
x_data=[]
y_data=[]
for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])
```

```python
import numpy as np
x_data, y_data=np.array(x_data),np.array(y_data)
```

```python
x_data, y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
5035-int(len(x_data)*0.7)
```

```python
splitting_len=int(len(x_data)*0.7)
x_train=x_data[:splitting_len]
y_train=y_data[:splitting_len]

x_test=x_data[splitting_len:]
y_test=y_data[splitting_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
##BUILDING THE NEURAL NETWORK 
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model=Sequential()
model.add(LSTM(128,return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size=1, epochs=2)
```

```python
model.summary()
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
inverse_predictions=scaler.inverse_transform(predictions)
```

```python
inverse_y_test=scaler.inverse_transform(y_test)
```

```python
inverse_predictions
```

```python
inverse_y_test
```

```python
rmse=np.sqrt(np.mean((inverse_predictions-inverse_y_test)**2))
```

```python
rmse
```

```python
plotting_data= pd.DataFrame(
{
    'original test data': inverse_y_test.reshape(-1),
    'predictions':inverse_predictions.reshape(-1)
}, index=google_data.index[splitting_len+100:]
)
plotting_data = plotting_data.sort_index(ascending=False)
plotting_data
```

```python
plot_graph((15,5), plotting_data, 'Test Data ')
```

```python
plot_graph((15,5), pd.concat([Adj_Close_Price[:splitting_len+100],plotting_data],axis=1),'Whole Data ')
```
